```python
import pandas as pd
movies=pd.read_csv('dataset1.csv')
```

```python
movies.head(10)
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies=movies[['id','title','genres','overview']]
```

```python
movies.columns
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genres']
```

```python
movies
```

```python
new_movies=movies.drop(columns=['overview','genres'])
```

```python
new_movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words="english")
```

```python
cv
```

```python
vector=cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_movies[new_movies['title']=="Spectre"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_movies.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_movies[new_movies['title']==movies].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_movies.iloc[i[0]].title)
    
    
```

```python
recommend("Batman Begins")
```
